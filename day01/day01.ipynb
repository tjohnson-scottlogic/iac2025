```python
# Load list of rotations
text_file = open("input.txt", "r")
lines = text_file.readlines()
text_file.close()

current_dial_position = 50
count_of_zeros = 0

for i in range (0,len(lines)):
  distance = int(lines[i][1:])

  if distance > 100:
    print(f"Distance is greater than 100, distance = {distance}")
    multiples = distance//100
    distance = distance - ((distance//100) * 100)
    print(f"Distance is now {distance}")
    count_of_zeros = count_of_zeros + multiples

  if(lines[i][0] == "L"):
    print("Left")
    print(lines[i][1:-1])
    current_dial_position = current_dial_position - distance
    print(f"current_dial_position is {current_dial_position}")
    if (current_dial_position < 0):
      current_dial_position = 100 + current_dial_position
      print(f"current_dial_position is {current_dial_position}")
    if (current_dial_position == 0):
      count_of_zeros = count_of_zeros + 1
      print(f"count_of_zeros = {count_of_zeros}")
    elif (current_dial_position + distance > 100):
      count_of_zeros = count_of_zeros + 1
      print(f"count_of_zeros = {count_of_zeros}")

  if(lines[i][0] == "R"):
    print("Right")
    print(lines[i][1:-1])
    current_dial_position = current_dial_position + distance
    print(f"current_dial_position is {current_dial_position}")
    if (current_dial_position > 99):
      current_dial_position = current_dial_position - 100
      print(f"current_dial_position is {current_dial_position}")
    if (current_dial_position == 0):
      count_of_zeros = count_of_zeros + 1
      print(f"count_of_zeros = {count_of_zeros}")
    elif (current_dial_position - distance < 0):
      count_of_zeros = count_of_zeros + 1
      print(f"count_of_zeros = {count_of_zeros}")

print(f"Final number for count_of_zeros is {count_of_zeros}")
```
