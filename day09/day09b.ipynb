```python
RED = 1
#FILENAME = "test.txt"
#FILENAME = "sample.txt"
FILENAME = "input.txt"
#WIDTH = 6
#WIDTH = 14
#WIDTH = 100000
#DEPTH = 6
#DEPTH = 9
#DEPTH = 100000

current_rectangle_area = 0
largest_rectangle_area = 0
```

```python
with open(FILENAME, "r") as location:
  locations = location.read().splitlines()
  print(locations)

for location in locations:
  print(f"location is {location}")
  location = location.split(",")
  print(f"x is {location[0]}")
  print(f"y is {location[1]}")
  for other_location in locations:
    other_location = other_location.split(",")
    print(f"other_location is {other_location}")
    current_rectangle_area = abs(((int(other_location[0]) - int(location[0])) +1) * ((int(other_location[1])- int(location[1]))+1))
    print(f"current_rectangle_area = {current_rectangle_area}")
    if (current_rectangle_area > largest_rectangle_area):
      largest_rectangle_area = current_rectangle_area

print(f"largest_rectangle_area = {largest_rectangle_area}")
```
